```python
import os
import openai

#input your api
openai.api_key = ''
```

```python
system_prompt = """You are a question-generating expert. If you are given information about Korean policy, you should create a question that can be answered through the information. However, the format of the question is a question that asks which users have a policy they can receive. Please create a "Korean" question.
'''''''''''
The folloinwg is some examples of the generated question for given passage.
Given Passage:
제목: 수원시 청년 주거 대출 프로그램
지원내용: 수원시에 거주하는 청년층이 받을 수 있는 대출 프로그램
연령: 만 19세 ~ 34세
거주지 및 소득: 수원시 거주, 중소기업에 재직 중인 청년
학력: 제한없음
전공: 제한없음
취업 상태: 중소기업에 1년 이상 재직 중
특화 분야: 제한없음

Korean Question:
수원시에 거주 중인 청년이고 중소기업에 1년 정도 다니고 있는 청년입니다. 제가 받을 수 있는 주거 관련 정책 프로그램이 있나요?

Given Passage:
제목: 유캔 두 코딩 인 강남
지원내용: 서울시 강남구에 거주 중인 취업 준비생에게 제공하는 코딩 교육 프로그램
연령: 만 19세 ~ 34세
거주지 및 소득: 서울시 강남구에 거주
학력: 대학교 졸업 이상
전공: 제한없음
취업 상태: 미취업
특화 분야: 제한없음

Korean Question:
코딩 관련 교육 프로그램에 대해 알려주세요.
"""
user_prompt = """Given Passage: {passage}\nKorean Question:"""
```

```python
def generate_question(passage):
    messages =[
        {'role':'system', 'content':system_prompt},
        {'role':'user', 'content':user_prompt.format(passage=passage)}
    ]
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=0.3,
        max_tokens=100,
        top_p=1.0,
        frequency_penalty=0.0,
        presence_penalty=0.0
    )

    return response['choices'][0]['message']['content']
```

```python
generated_question = generate_question(content)
```
